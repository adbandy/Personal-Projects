# Visualizing The Gender Gap In College Degrees

In this visualisation project we will be exploring the differences between Men and Women in the database from The Department of Education Statistics containing the percentage of bachelor's degrees granted to women from 1970 to 2012.

### Load In The Data

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

### Creating A Grid Of Subplot 

In this project we will show the changes overtime in the share of Men and Women doing each degree.

First lets create some lists to aid us:

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

Next is the figure creation which is complicated but will be aided with code comments

```python
#Set Colours to use for the lines in the graphs#
#The numbers are in the RGB format#
#so 0/255, 255/255, 0/255 represents Green only#
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

#Set the plot size#
fig = plt.figure(figsize=(16, 20))
#First Plot using stem degrees in the first subplot column.#
for sp in range(0,17,3):
#The graphs will be placed in position 1,4,7,10,13,16 to display as the left column#
    ax = fig.add_subplot(6,3,sp+1)
#Find the index in the stem_list by rounding the position divided by 3#
    cat_index = int(sp/3)
#Select the two lines to plot#
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
#Give it a clean look with no spines#
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
#Set the ranges of the graph and the tick spacings#
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
#Set the title and add a line showing the 50% share of Women:Men#
    ax.set_title(stem_cats[cat_index])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha = 0.3)
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom= "off")
    if cat_index == 5:
        ax.tick_params(labelbottom="on")
#Show the line labels on the first and last graphs in the column#
    if cat_index == 0:
        ax.text(2005, 8, 'Women')
        ax.text(2003, 87, 'Men')
    elif cat_index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 8, 'Women')
        
#The rest of the code below mostly is a repeat of above to generate the other two columns#

#Second Plot using liberal arts degrees in the second subplot column.   
for sp in range(1,15,3):
    ax = fig.add_subplot(6,3,sp+1)
    cat_index = int((sp-1)/3)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255) , alpha=0.3)
    ax.set_title(lib_arts_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom= "off")    
    if cat_index == 4:
        ax.tick_params(labelbottom="on")
    if cat_index == 0:
        ax.text(2005, 20, 'Men')
        ax.text(2003, 80, 'Women')
# This function is omitted as only 5 degrees in this catagory.
#     elif cat_index == 5:
#         ax.text(2006, 87, 'Men')
#         ax.text(2005, 8, 'Women')

#Third plot using 'other' degrees in the third subplot column.
for sp in range(2,19,3):
    ax = fig.add_subplot(6,3,sp+1)
    cat_index = int((sp-2)/3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom= "off")
    if cat_index == 5:
        ax.tick_params(labelbottom="on")
    if cat_index == 0:
        ax.text(2005, 5, 'Men')
        ax.text(2003, 90, 'Women')
    elif cat_index == 5:
        ax.text(2005, 60, 'Men')
        ax.text(2003, 35, 'Women')

#Display the graph. Then save it locally#
plt.show()
plt.savefig("gender_degrees.png")
```

### Observations

- The percentage of Women and Men in degrees have been converging over the past 40 years in general.

Very few degrees showed majority of students as Female in 1970. These were:
- Foreign Languages
- English
- Art and Performance
- Health Professions
- Public administration
- Education

Many degrees show a flip from a Male dominated degree to a Female dominated degree over time:
- Psychology - Women have consistently increased as a percentage here since 1980
- Biology - Around 1990 the flip occured. However the gap has been closing (to more 50:50 split) over the last 5 years of this dataset
- Communication and Journalism - The flip occurred before 1980s and a significant gap remains in 2010
- Social Sciences and History - This has flipped twice over the period but essentially the flips are when the difference is small so this degree is essentially balanced in representation.
- Business - The same trend as the Social Sciences is seen here except the rate of convergence is far greater

Some degrees have nearly no more balance than they did in the 1970s:
- Computer Science - There was a period in the early 1980s where convergence looked likely but this has since diverged.
- Psychology - In the 1970s this was close to balanced. Now it is disproportionately dominated by Men
- English - There has been very gradual divergence in this overal.
- Communication and Journalism - The size of the disparity remains but has flipped from Male to Female dominated
- Arts and Performance - No significant change in the gender disparity
- Health Professionals & Public Administration & Education - Divergence has increased over time and are disproportionately dominated by Women

Some degrees are now essentially balanced in their Male:Female representation
- Social Sciences and History, Agriculture and Business

It is of note that these are merely correlations and say nothing as to the causations (the why).
